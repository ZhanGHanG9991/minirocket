```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
insectSound = [1.15590981, 3.113912788, 13.08650136, 64.21685102, 287.3746]
fruitFlies = [3.007717321, 6.460926256, 18.0783085, 77.79989341, 369.832764]
mosquitoSound = [2.582793489, 5.207610789, 15.30720129, 76.57430078, 247.6187271, 473.5597125, 694.8291147, 967.544089]
ducksAndGeese = [0.21192739, 0.329376039, 0.663334629, 1.322438206, 2.899269552, 5.854076017, 14.44972038, 34.23396049]
```

```python
# Figure6_1
x = [i for i in range(10, 15)]
fig = plt.figure(figsize=(9,9))
# 为图像添加子目标ax2,后续所有的操作为在ax2上操作
ax2  = fig.add_subplot()
ax2.plot(x, fruitFlies, "-o")
ax2.set_xlim(9, 18)
ax2.set_xlabel('Training Set Size')
ax2.set_ylim(10**-1, 10**4)
ax2.set_yscale('log')
ax2.set_ylabel('Training Time(s)')
ax2.set_title('FruitFlies')
plt.savefig('fruitFlies.jpg', dpi=400)
```
